```python
# Data Preprocessing

# Separate features and target variable

X= df.drop(["id",'Depression'],axis=1)
Y= df[['Depression']]
# Identify numerical and categorical columns
numerical_columns = [  "Age", "Academic Pressure",'Work Pressure',"CGPA","Study Satisfaction","Job Satisfaction","Work/Study Hours",
                     "Avarage_satisfactiont","Financial Stress"]
categorical_cols = ['Gender','City','Profession',"Have you ever had suicidal thoughts ?","Family History of Mental Illness"]

ordinal_cols = ['Dietary Habits','Sleep Duration','Degree','Have you ever had suicidal thoughts ?',
            "Family History of Mental Illness"]
# Preprocessing for numerical data: impute missing values and scale features
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ("poly",PolynomialFeatures(degree=5,include_bias=False)),
    ('scaler', StandardScaler())
])

# Preprocessing for categorical data: impute missing values and one-hot encode
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore')),
    
])

ordinal_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ("ordinal", OrdinalEncoder(categories=ordinal_categories))
])
# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_columns),
        ('cat', categorical_transformer, categorical_cols),
        ('ordinal',ordinal_transformer,ordinal_cols)
    ])
preprocessor_ploy = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_columns),
        ('cat', categorical_transformer, categorical_cols),
        ('ordinal',ordinal_transformer,ordinal_cols)
    ])

# Split the data into training and testing sets
X_train , X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, shuffle=True, random_state=42, stratify=Y)
X_train['Avarage_satisfactiont'] = (X_train['Job Satisfaction'] + X_train['Study Satisfaction'])/2
X_test['Avarage_satisfactiont'] = (X_test['Job Satisfaction'] + X_test['Study Satisfaction'])/2
X_train['Avarage_Pressure'] = (X_train['Work Pressure'] + X_train['Financial Stress'])/2
X_test['Avarage_Pressure'] = (X_test['Work Pressure'] + X_test['Financial Stress'])/2


# Apply the preprocessing pipeline to the training data
X_train_preprocessed_poly = preprocessor.fit_transform(X_train)
X_test_preprocessed_poly = preprocessor.transform(X_test)
preprocessor_ploy
```

```python
# Model Building
from sklearn.calibration import cross_val_predict


log_reg = LogisticRegression(random_state=42, C=1.0, max_iter=520, warm_start=True)
log_reg.fit(X_train_preprocessed_poly, y_train)

y_train_pred = log_reg.predict(X_train_preprocessed_poly)

y_valid_pred = cross_val_predict(log_reg, X_train_preprocessed_poly, y_train, cv=3)
precision = precision_score(y_train,y_valid_pred) # == TP/TP+FP =>
recall = recall_score(y_train,y_valid_pred) # == TP/TP+FN =>
f1_score = f1_score(y_train,y_valid_pred)
print(f"Train Accuracy: {accuracy_score(y_train, y_train_pred)}")
print(f"Validation Accuracy: {accuracy_score(y_train, y_valid_pred)}")

print(f"the valid precision :{precision}")
print(f"the valid recall :{recall}")
print(f"the valid f1_score :{f1_score}")
```
