```python
# Manpulate
import numpy as np
import pandas as pd

# Visualization
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('Student Depression Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
## describtion for categorical columns
for col in df.select_dtypes(include=['object']).columns:
    print(f"Column: {col}")
    print(df[col].value_counts())
    print('____________________________')
```

```python
for col in df.select_dtypes(include=['object']).columns:
    print(f"Column: {col}")
    print(df[col].value_counts())
    
    # Plotting Bar Plot for Categorical Data
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, palette='viridis')
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Univariate Analysis for Numerical Columns
num_cols = df.select_dtypes(include="number").columns

for col in num_cols:
    fig, axes= plt.subplots(nrows=1, ncols=2)
    sns.histplot(df[col], kde=True, ax=axes[0])
    sns.boxplot(x=df[col], ax=axes[1])
    plt.show()
```

```python
Male = df.loc[df["Gender"] == "Male"].count()[0]
Female = df.loc[df["Gender"] == "Female"].count()[0]
labels =  ["Male", "Female"]
plt.pie([Male, Female],
    labels = labels,
    autopct='%1.1f%%',  # Display percentages
    startangle=90,      # Rotate pie chart
)

plt.title('Male Female Pie Chart')  # Add title
plt.show()
```

```python
healthy = df.loc[df["Dietary Habits"] == "Healthy"].count()[0]
unhealthy = df.loc[df["Dietary Habits"] == "Unhealthy"].count()[0]
moderate = df.loc[df["Dietary Habits"] == "Moderate"].count()[0]
Other = df.loc[df["Dietary Habits"] == "Other"].count()[0]
labels =  ["Healthy", "Unhealthy", "Moderate", "Other"]
plt.pie([healthy, unhealthy, moderate, Other],
    labels = labels,
    autopct='%1.1f%%',  # Display percentages
    startangle=90,      # Rotate pie chart
)

plt.title('Dietary Habits Pie Chart')  # Add title
plt.show()
```

```python
counter = 1
columns=['Gender','Dietary Habits','Sleep Duration','Have you ever had suicidal thoughts ?',
            "Family History of Mental Illness",'Degree']
plt.figure(figsize=(20,300))
for column in columns:
    plt.subplot(40,1,counter)
    counter+=1
    sns.countplot(x='Depression',hue=column ,data = df)
```

```python
numerical_columns = [  "Age", "Academic Pressure",'Work Pressure',"CGPA","Study Satisfaction","Job Satisfaction","Work/Study Hours",
                     "Avarage_satisfactiont","Financial Stress"]
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Academic Pressure', data=df)

# Set labels and title for the plot
plt.xlabel('age')
plt.ylabel('Academic Pressure')
plt.title('Correlation between Academic Pressure  and age')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Financial Stress', data=df)

# Set labels and title for the plot
plt.xlabel('age')
plt.ylabel('Financial Stress')
plt.title('Correlation between Financial Stress  and age')

# Display the plot
plt.show()
```

```python
numerical_data = df.select_dtypes(include=['number'])
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_data.corr(), annot=True, cmap='viridis', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
